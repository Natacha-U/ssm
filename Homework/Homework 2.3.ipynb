### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi y)\\
u(1,y) = \sin (\pi y)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

$$ u(i,j=0) = \sin(2\pi x)\\
u(i,j=1) = \sin(2\pi x)\\
u(i=0,j) = \sin(3\pi y)\\
u(i=1,j) = \sin(\pi y)$$

```python
%%time 

# Set up the matrix, say it is 50 x 50
x, y = 100,100

# Set up the grid for the 
u = np.zeros((x, y))
# X, Y = np.meshgrid(np.arange(0, x), np.arange(0, y))
# colourMap = plt.cm.hot

# Set up the boundary conditions for the 4 edges

# u(i , j = 0) is the left edge, index = 0 
# u(i , j = 1) is the right edge, index = 50 

for i in range(x):
    u[i][0] = np.sin(2*np.pi*i/x)
    u[i][-1] = np.sin(2*np.pi*i/x)
    
# u(i = 0 , j) is the top edge
# u(i = 1 , j) is the bottom edge

for j in range(y):
    u[0][j] = np.sin(3*np.pi*j/y)
    u[-1][j] = np.sin(np.pi*j/y)
    
# Now we get a matrix with values on the edges but empty in the center, so we will have to compute 
# u(i,j) and work our way towards the center 
# We start the index from 1 because we already made the edges
for k in range(2500): # Repeatttt 555
    for i in range(1, x-1):
        for j in range(1, y-1):
            u[i][j] = (1/4) * (u[i+1][j] + u[i-1][j] + u[i][j+1] + u[i][j-1])
```

```python
%%time 
plt.figure(figsize = (6,6))
plt.xlim(0,x-1)
plt.ylim(0,y-1)
plt.imshow(u, cmap = 'hot')
plt.colorbar()
```
